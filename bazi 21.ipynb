```python
def random_chose():
    x=random.choice(kart)
    for i in kart:
        if i==x:
            kart.remove(i)
    return x
```

```python
def dadan_kart_pc():
    global kart_pc
    kart_pc=[]
    for i in range(2):
        kart_pc.append(random_chose())
    return kart_pc
```

```python
def dadan_kart_bazikon():
    global kart_bazikon
    kart_bazikon=[]
    for i in range (2):
        kart_bazikon.append(random_chose())
    return kart_bazikon
```

```python
def kart_khastan_bazikon():
    halghe=True
    global mablagh
    global kart_bazikon
    while halghe==True:
        u=input('aya mikhahid kart bekeshid, Yes or No ? ')
        u=u.lower()
        if u=='yes':
            clear_output()
            kart_bazikon.append(random_chose())
            print('kart ha: ',kart_bazikon)
            print('=============================') 
            print('jam emtiyaz: ',jam_emtiyaz_bazikon())
            if bakhtan_bazikon()==False:
                clear_output()
                print('jam emtiyaz shoma bishtar az 21 ahod')
                print('=============================')
                moghayese_bordan()
                halghe=False
        elif u=='no':
            clear_output()
            kart_khastan_pc()
            moghayese_bordan()
            halghe=False
        else:
            print('lotfan bein Yes va No entekhab konid')
            
    print('=============================')    
    print('kart haye bazikon: ' , kart_bazikon)
    print('=============================')
    print('jam emtiyaz bazikon: ' , jam_emtiyaz_bazikon())
    print('=============================')
    print('kart haye pc: ' , kart_pc)
    print('=============================')
    print('jam emtiyaz pc: ' , jam_emtiyaz_pc())
    print('=============================')
    print('mablagh shart bandi: ' , mablagh)
    print('=============================')
    print('mablagh kol sarmaye baghi mande: ' , mohasebat())
    
            
    
```

```python
def jam_emtiyaz_bazikon():
    global kart_bazikon
    emtiyaz_bazikon=0
    emtiyazz=0
    jaz=0
    for i in range(len(kart_bazikon)):
        emtiyazz+=emtiyaz_kartha[kart_bazikon[i][1]]
        
        if kart_bazikon[i][1]=='1':
            emtiyaz_bazikon+=emtiyaz_kartha[kart_bazikon[i][1]]
            if emtiyaz_bazikon>21:
                emtiyaz_bazikon-=emtiyaz_kartha[kart_bazikon[i][1]]
                emtiyaz_bazikon+=1
            
        elif emtiyazz>21:
            for j in range(len(kart_bazikon)):
                if kart_bazikon[j][1]=='1':
                    if jaz==0:
                        emtiyaz_bazikon-=emtiyaz_kartha[kart_bazikon[j][1]]
                        jaz=1
                        emtiyaz_bazikon+=1
            emtiyaz_bazikon+=emtiyaz_kartha[kart_bazikon[i][1]]
            emtiyazz=emtiyaz_bazikon
        
        
        else:
            emtiyaz_bazikon+=emtiyaz_kartha[kart_bazikon[i][1]]   
            
    return(emtiyaz_bazikon)
    
```
